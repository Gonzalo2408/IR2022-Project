```python
from pandas import *
 
# reading CSV file
data = read_csv("/content/data/cord19_corpus/2020-07-16/metadata.csv")
```

```python
data['sha']
```

```python
corpus = data[['sha', 'abstract']]
new_corpus = {}

for i in range(len(corpus)):

  new_corpus[corpus['sha'][i]] = corpus['abstract'][i]  
```

```python
len(new_corpus)
```

```python
import json

with open('/content/data/cord19_corpus/document_parses/new_corpus.json', 'w') as f:
  json.dump(new_corpus, f)
```
